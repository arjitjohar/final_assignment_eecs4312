```python

from google_play_scraper import Sort, reviews_all



#get all package names from MyAppDecsription.csv
df = pd.read_csv('MyAppDescriptions.csv')
package_names = df['Package Name'].tolist()


for package_name in package_names:
  #remove any character before the & sign
  package_name = package_name.split('&')[0]
  print(package_name)




  

  result = reviews_all(
    package_name,
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    #number of reviews to get
    count=1000,
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
  )


  df_busu = pd.DataFrame(np.array(result),columns=['review'])
  df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))
  df_busu = df_busu[['userName','content','score']]
  df_busu.columns =['Reviewer name', 'Review','Rating']
  df_busu.insert(0,'Package name','')
  df_busu['Package name'] = package_name
  file_name = (package_name+'.csv')
  df_busu.to_csv(file_name)









 


    
```

```python

import glob
import os
path = os.getcwd()
import pandas as pd
extension = 'csv'
os.chdir(path)
result = glob.glob('*.{}'.format(extension))

#remove the MyAppDescriptions.csv file from the list
result.remove('MyAppDescriptions.csv')

#remove MyAppFeatures.csv file from the list
result.remove('MyAppFeatures.csv')

print(result)



## similiar to lab 2 processing of text


df = pd.DataFrame()
for file in result:
    df = pd.read_csv(file)
    
    

    review = df['Review']


    
    for i in range(len(review)):


        review[i] = str(review[i])



        #remove punctuation
        review[i] = review[i].replace('[^\w\s]','')

        #remove special characters and emojis and other symbols
        review[i] = re.sub('[^A-Za-z0-9]+', ' ', review[i])

        #if it is a number or a float, convert it to a word
        review[i] = re.sub(r'\d+', lambda x: num2words.num2words(int(x.group(0))), review[i])

        #remove extra white spaces 
        review[i] = review[i].replace('\s+', ' ')

        #turn all words into lower case
        review[i] = review[i].lower()

        #remove stop words
        stop_words = set(stopwords.words('english'))

        review[i] = [word for word in review[i].split() if word not in stop_words]

        #lemmatize the words
        lemmatizer = WordNetLemmatizer()
        review[i] = [lemmatizer.lemmatize(word) for word in review[i]]

        #join the words back together
        review[i] = ' '.join(review[i])        


        
        

        

        #write the review to a csv file
        
        df['Review'][i] = review[i]

        df_new = df[['Package name','Reviewer name','Review','Rating']]


        


    #save to csv file with the same name as the original csv file and add _cleaned to the end of the file name
    df_new.to_csv(file[:-4]+'_cleaned.csv')


    
    









```

```python
from textblob import TextBlob


#make dataframe with the sentiment analysis of each review from all the _cleaned.csv files

dfs = []
df = pd.DataFrame()

for file in result:
    if file.endswith('_cleaned.csv'):
        df = pd.read_csv(file)
        df['Polarity'] = df['Review'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
        df = df[['Package name','Review','Polarity']]

        dfs.append(df)
        

      
    else:
        continue


#show first 15 rows of the dataframe
df_textBlob_sentiment = pd.concat(dfs)
df_textBlob_sentiment.head(15)





```

```python
#import vader sentiment analysis
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

#make dataframe with the sentiment analysis of each review from all the _cleaned.csv files

dfs = []
df = pd.DataFrame()

for file in result:
    if file.endswith('_cleaned.csv'):
        df = pd.read_csv(file)
        df['vader'] = df['Review'].apply(lambda x: analyzer.polarity_scores(str(x)))
        df = df[['Package name','Review','vader']]

        dfs.append(df)
        

      
    else:
        continue
#show first 15 rows of the dataframe
df_vader_sentiment = pd.concat(dfs)

#show polarity score for each review
df_vader_sentiment.head(15)
```

## How does the sentiments retrieved by Textblob and Vader compare with each other? How do you interpret the similarity/difference? Which one is the best option for review analysis of your apps? Why?



```python
#compare the two sentiment analysis methods
df_vader_sentiment['vader'] = df_vader_sentiment['vader'].apply(lambda x: x['compound'])
df_vader_sentiment.head(15)

df_textBlob_sentiment.head(15)

#merge the two dataframes

df_sentiment = pd.merge(df_vader_sentiment, df_textBlob_sentiment, on=['Package name','Review'])

#change Polarity column name to either positive neutral or negative
df_sentiment['Polarity_check'] = df_sentiment['Polarity'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))

#change vader column name to either positive neutral or negative


#change the vader column name to either positive neutral or negative
df_sentiment['vader_check'] = df_sentiment['vader'].apply(lambda x: 'positive' if x >= 0.05  else ('negative' if x <= -0.05 else 'neutral'))




df_sentiment.head(15)

#save the dataframe to a csv file
df_sentiment.to_csv('sentiment_analysis.csv')

#show the first 15 rows of the dataframe

df_sentiment.head(15)
```

```python
#show numbe of positive, negative and neutral reviews for each vader and textblob
df_vader_group = df_sentiment.groupby('Package name')['vader_check'].value_counts()

df_text_group = df_sentiment.groupby('Package name')['Polarity_check'].value_counts()


#compare the number of positive negative and neutral reviews for each sentiment analysis method


df_vader_group

df_text_group









```
